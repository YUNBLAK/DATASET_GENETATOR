```python
import pandas as pd
import random
import matplotlib.pyplot as plt

df = pd.read_csv("regression.csv")
x = df["X"]
y = df["Y"]

newx = []
newy = []

print("DONE")
for i in range(0, len(x)):
    for j in range(0, 10000):
        newx.append(x[i] - random.uniform(-10, 10))
        newy.append(y[i] - random.uniform(-10, 10))
print("DONE")

plt.figure(figsize = (15, 4))
plt.scatter(newx, newy, s = 0.05)
plt.show()

dict = {"X" : newx,
       "Y" : newy}

newdf = pd.DataFrame(dict)
newdf.to_csv("BIGDATAz.csv", index = False)
```
